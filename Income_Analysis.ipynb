# **MACHINE LEARNING**

```python
#our main aim is to create a new entry and predict its income class with models we trained
import numpy as np
```

```python
#transforming categorical columns into numeric ones
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
#before label encoding
data["WORK CLASS"].head()
```

```python
#after label encoding
data["WORK CLASS"] = labelencoder.fit_transform(data["WORK CLASS"])
data["WORK CLASS"].head()
```

```python
#applying label encoder for remaining categorical columns
data["EDUCATION"] = labelencoder.fit_transform(data["EDUCATION"])
data["MARTIAL STATUS"] = labelencoder.fit_transform(data["MARTIAL STATUS"])
data["OCCUPATION"] = labelencoder.fit_transform(data["OCCUPATION"])
data["NATIVE COUNTRY"] = labelencoder.fit_transform(data["NATIVE COUNTRY"])
data["INCOME"] = labelencoder.fit_transform(data["INCOME"])
```

```python
data.head()
```

```python
#generate heatmap by using seaborn
import seaborn as sns
sns.heatmap(data = data.corr(),annot = True)
```

```python
#setting up 'X' and 'Y'
x = data[['AGE', 'WORK CLASS', 'EDUCATION','MARTIAL STATUS', 'OCCUPATION', 'HOURS-PER-WEEK', 'NATIVE COUNTRY']]
y = data[['INCOME']]
```

```python
#train and test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)  #30% of our data is going to test set
```

```python
#since x values have different ranges, we will scale them by using standardscaler from sklearn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.fit_transform(x_test)
```

```python
#defining a function to get accuracy score
from sklearn.metrics import accuracy_score
def modelresults(predictions):    #"predictions" are the values we are going to predict with our trained model
                                  #predictions will be made on test set
  print(f"accuracy score of the model is {accuracy_score(y_test,predictions)}")
```

```python
#training our models with logistic regression
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(scaled_x_train, y_train)
```

```python
#making prediction with our model using Linear Regression model
log_predictions =  logmodel.predict(scaled_x_test)
modelresults(log_predictions)
```

```python
#training KNN model
#we will use grid search for hyperparameter tuning
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
knn = KNeighborsClassifier()
k_values = list(range(1, 30))
pipe = Pipeline(steps = [('knn', knn)])
param_grid = {"knn__n_neighbors": k_values}
cv_classifier = GridSearchCV(pipe, param_grid, cv = 5, scoring = "accuracy")
cv_classifier.fit(scaled_x_train, y_train)
```

```python
#make predictions using KNN model
pred_gridknn = cv_classifier.predict(scaled_x_test)
modelresults(pred_gridknn)
```

```python
#training SVM model and we also use grid search for hyperparameter tuning
from sklearn.svm import SVC
svm = SVC()
param_grid_svr = {"C": [0.1,0.5,1], "kernel": ["linear","rbf", "poly"]}
gridsvr = GridSearchCV(svm,param_grid_svr)
gridsvr.fit(scaled_x_train, y_train)
```

```python
#make predictions using SVM model
pred_svm = gridsvr.predict(scaled_x_test)
modelresults(pred_svm)
```

```python
#training Random Forest model and we also use grid search for hyperparameter tuning
from sklearn.ensemble import RandomForestClassifier
rfr_model = RandomForestClassifier()
n_estimators = [32,64,128]
max_features = [2,3,4,5]
bootstrap = [True, False]
oob_score = [True, False]
param_grid_rfr = {"n_estimators": n_estimators, "max_features": max_features, "bootstrap": bootstrap, "oob_score": oob_score}
grid_rfr = GridSearchCV(rfr_model,param_grid_rfr)
grid_rfr.fit(scaled_x_train, y_train)
```

```python
pred_rfr =  grid_rfr.predict(scaled_x_test)
modelresults(pred_rfr)
```
